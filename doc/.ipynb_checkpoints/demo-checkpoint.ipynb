Next, make sure you have your SimaPro csv file ready to go! So in SimaPro, File -> Export and choose the same options as in the screenshot below:

Well now we just use Simaporter which will do all the work for us.

- *bw_project_name*: the name of the brightway2 project into which the simapro project will be installed
- *ecoinvent_db_name_in_bw*: the name of the ecoinvent database with which the simapro project will be linked, e.g., "ecoinvent3.6 cut-off"
- *biosphere_db_name_in_bw*: the name of the biosphere database with which the simapro project will be linked, by default the name should be "biosphere3"
- *sp_csv_file*: the path to the simapro csv file containing all the information from the simapro project
- *db_name*: the name which the importer project of simapro will have
- *delimiter*: the delimiter which was used in the simapro csv file

```python
self = simporter.Simporter(bw_project_name='ecoinvent3.6',
                ecoinvent_db_name_in_bw='ecoinvent3.6 cut-off',
                biosphere_db_name_in_bw='biosphere3',
                sp_csv_file='C://Users/Maxime/Desktop/Pokeballs-LCA.csv',
                db_name='Pokeballs',
                delimiter=';')
```
